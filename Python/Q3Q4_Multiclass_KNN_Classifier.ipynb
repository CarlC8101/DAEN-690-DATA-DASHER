```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
cleanedset = pd.read_csv('C:/Users/sampa/OneDrive/Desktop/GMU/Spring_2021/Dataset/cleandata_q3q4.csv')
cleanedset.head(5)
```

```python
cleanup_nums = {"Shift": {"A - Shift": 1, "B - Shift": 2, "C - Shift" : 3},
               "PatientOutcome": {"Treated & Transported ": 1, "Standby (No Services Performed)": 2, 
                                   "Treated, Transferred Care ": 3, 
                                   "Standby (Operational Support Provided)": 4, 
                                  "Patient Refusal  (AMA)": 5, "Patient Dead at Scene (No EMS CPR)": 6,
                                   "Patient Dead at Scene (EMS CPR Attempted)": 7, "No Treatment/Transport Required": 8,
                                  "No Patient Found": 9, "EMS Assist (Other Agency)": 10, "Canceled (Prior to Arrival) ": 11,
                                  "Canceled (On Scene, No Patient Contact) ": 12}
}

cleanedset = cleanedset.replace(cleanup_nums)
cleanedset.head(5)
```

```python
df = cleanedset[['Shift', 'FireStation', 'tenure', 'PatientOutcome']]
```

```python
df.head(5)
```

```python
#Data Normalisation
x_data = df.drop(['PatientOutcome'],axis=1)
y_data = df['PatientOutcome']
MinMaxScaler = preprocessing.MinMaxScaler()
X_data_minmax = MinMaxScaler.fit_transform(x_data)
data = pd.DataFrame(X_data_minmax,columns=['Shift', 'FireStation', 'tenure'])
df.head()
```

```python
x_data.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y_data,test_size=0.2, random_state = 1)
knn_clf=KNeighborsClassifier()
knn_clf.fit(X_train,y_train)
ypred=knn_clf.predict(X_test) #These are the predicted output values
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, ypred)
print("Confusion Matrix: ")
print(result)
result1 = classification_report(y_test, ypred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,ypred)
print("Accuracy:",result2)
```
