```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('genderclean.csv')
```

```python
cleanedset = df
```

<font size="5">Split entire dataset into three sections</font> which are Maleinteraction, Femaleinteraction, Mixedinteraction

```python
cleanedset['Maleinteraction'] = 0
for i, row in cleanedset.iterrows():
    cleanedset.at[i,'Maleinteraction'] = 1 if row['PatientGender'] == 'Male' and row['FRDPersonnelGender'] == 'Male' else 0
```

```python
cleanedset['Femaleinteraction'] = 0
for i, row in cleanedset.iterrows():
    cleanedset.at[i,'Femaleinteraction'] = 1 if row['PatientGender'] == 'Female' and row['FRDPersonnelGender'] == 'Female' else 0
```

```python
cleanedset['Mixedinteraction'] = 0
for i, row in cleanedset.iterrows():
    cleanedset.at[i,'Mixedinteraction'] = 1 if row['Femaleinteraction'] == 0 and row['Maleinteraction'] == 0 else 0
```

```python
male = cleanedset.loc[cleanedset['Maleinteraction'] == 1]
female = cleanedset.loc[cleanedset['Femaleinteraction'] == 1]
mix = cleanedset.loc[cleanedset['Mixedinteraction'] == 1]
```

<font size="5">For male interaction: </font>

```python
contingenc_table = pd.crosstab(male['Medication_Given_Description'],
                              male['Procedure_Performed_Description'],
                              margins = True)
contingenc_table
```

Chi-squared test

```python
arr = []
for i in range(len(contingenc_table)-1):
    arr += [contingenc_table.iloc[i][0:-1].values]

f_obs = np.array(arr)
stats.chi2_contingency(f_obs)[0:3]
```

CarmerV

```python
X2 = stats.chi2_contingency(f_obs)[0]
n = np.sum(f_obs)
minDim = min(f_obs.shape)-1
V = np.sqrt((X2/n)/minDim)
print(V)
```
